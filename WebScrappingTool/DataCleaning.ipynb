# Remove $ and USD from the current_index_price, and other columns

```python
# df["current_index_price"] = df["current_index_price"].str.strip("$")
# df["Previous close"] = df["Previous close"].str.strip("$")

temp=[]
for col in df.select_dtypes("object").columns:
    if df[col].str.contains("\$").all():
        # print("yes", col)
        df[col]=df[col].str.replace("\$","", regex = True)
        df[col]=df[col].apply(lambda x:float(x))
        temp.append(col)

df["Market cap"] = df["Market cap"].str.strip(" USD")
```

```python
df[temp]
```

## Remove %

```python
temp=[]
for col in df.select_dtypes("object").columns:
    if df[col].str.contains("%").any():
        
        df[col]=df[col].str.strip("%")
        df.rename(columns = {f"{col}":f"{col} %"}, inplace = True)
        temp.append(f"{col} %")
```

```python
df[temp]
```

## Remove M and B from columns 

- Find columns with Bs

```python
billions=[]
for col in df.select_dtypes("object").columns:
    # print(col)
    if df[col].str.contains("B").all():
        billions.append(col)
# billions

if len(billions):
    print(billions)
    for col in billions:
        # df[col]=df[col].apply(lambda x:x.replace("B",""))
        df[col]=df[col].str.strip("B")
        df.rename(columns = {f"{col}":f"{col} ($B)"}, inplace = True)
```

- Remove Ks

```python
Ks=[]
for col in df.select_dtypes("object").columns:
    # print(col)
    if df[col].str.contains("K").all():
        Ks.append(col)
# billions
if len(Ks):
    print(Ks)
    for col in Ks:
        df[col]=df[col].str.strip("K")
        df[col]=df[col].apply(lambda x:float(x))
        df.rename(columns = {f"{col}":f"{col} in ($K)"}, inplace = True)
```

- Remove Ms

```python
millions=[]
for col in df.select_dtypes("object").columns:
    # print(col)
    if df[col].str.contains("M").all():
        millions.append(col)
# billions
if len(millions):
    print(millions)
    for col in millions:
        df[col]=df[col].str.strip("M")
        df[col]=df[col].apply(lambda x:float(x))
        df.rename(columns = {f"{col}":f"{col} in ($M)"}, inplace = True)
```

```python
trillions=[]
for col in df.select_dtypes("object").columns:
    # print(col)
    if df[col].str.contains("T").all():
        trillions.append(col)
# billions
if len(trillions):
    print(trillions)
    for col in trillions:
        df[col]=df[col].str.strip("T")
        df[col]=df[col].apply(lambda x:float(x))
        df.rename(columns = {f"{col}":f"{col} in ($T)"}, inplace = True)
```

## columns with M, B, and T

```python
temp=[]
for col in df.select_dtypes("object").columns:
    # print(col)
    if df[col].str.contains("K|M|B|T").any():
        temp.append(col)
temp
```

```python
df[temp]
```

- Fix Market cap, Net income, Shares outstanding, Cash from operations, investing, and financing, Net change in cash, and free cash flow

```python
print(temp)
for rem in ['Company','CEO','HQ City', 'HQ State']:
    temp.remove(rem)
print(temp)
```

```python
df[temp]
```

```python
for col in temp:
    rows_k = df[col].index[df[col].str.contains("K").values]
    rows_b = df[col].index[df[col].str.contains("B").values]
    rows_m = df[col].index[df[col].str.contains("M").values]
    rows_t = df[col].index[df[col].str.contains("T").values]

    if len(rows_k):
        rows=rows_k
        df.loc[rows,col]=df.loc[rows,col].apply(lambda x:x.replace("K",""))
        df.loc[rows,col]=df.loc[rows,col].apply(lambda x:float(x)/1000)  # K TO M

    if len(rows_b):
        rows=rows_b
        df.loc[rows,col]=df.loc[rows,col].apply(lambda x:x.replace("B",""))
        df.loc[rows,col]=df.loc[rows,col].apply(lambda x:float(x)*1000)  # B TO M

    if len(rows_m):
        rows=rows_m
        df.loc[rows,col]=df.loc[rows,col].apply(lambda x:x.replace("M",""))
        df.loc[rows,col]=df.loc[rows,col].apply(lambda x:float(x)*1)  

    if len(rows):
        rows=rows_t
        df.loc[rows,col]=df.loc[rows,col].apply(lambda x:x.replace("T",""))
        df.loc[rows,col]=df.loc[rows,col].apply(lambda x:float(x)*1000000) # T to M 
    
    df.rename(columns = {f"{col}":f"{col} ($M)"}, inplace = True)
```

# Change the date format, and keep only the founding year instead of month and day

```python
df["date_fetched"]=pd.to_datetime(df["date_fetched"])
df["FinReported"]=pd.to_datetime(df["FinReported"])
df["Founded"]=pd.to_datetime(df["Founded"])
```

```python
df["Founded"]=df["Founded"].dt.strftime('%Y')
df.rename(columns = {"Founded":"Year founded in"}, inplace = True)

df["date_fetched"]=df["date_fetched"].dt.strftime('%d-%m-%Y')
df["FinReported"]=df["FinReported"].dt.strftime('%d-%m-%Y')
```
